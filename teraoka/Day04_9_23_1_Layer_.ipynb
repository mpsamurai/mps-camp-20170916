```python
# 課題の自分のコード（bitデータ(1を与えるとy20は1を返し、y21は反転を返す）
import numpy as np
class Layer:
    
    def __init__(self, m, n): 
        self._rate = 0.01
        self._weight = np.random.rand(m, n) - 0.5  # m行n列を設定(-0.5をすることで-0.5 ~ 0.5の範囲にする)
        self._bias = np.zeros((m, 1))  # m行1列の　0ベクトルを作成してくれている  # バイアスは出力の行
        self._y = None
        self._delta = None
        
    def set_state(self, x):
        """入力値を内積"""
        s = (self._weight.dot(x))+1*self._bias
        return s

    def activate(self, s):
        "今回はシグモイド関数"
        self._y = 1 / (1 + np.exp(-s))
        return self._y
    
    def foward(self, x):
        """順伝播"""
        return self.activate(self.set_state(x))
    
    def back_propagation(self, t):
        """勾配計算(誤差逆伝播)"""
        de = self._y - t
        dy = self._y * (1 - self._y) 
        self._delta = de * dy
        return self._delta
    
    def update(self, x):
        """パラメーター更新"""
        self._weight -= self._rate * self._delta * x.T  # 入力値(y10を与えている)
        self._bias -= self._delta * self._bias 
        return self._y
    
    def error_function(self, y, o):
        return np.power(y[0] - o[0],2) + np.power(y[1] - o[1],2)

if __name__ == '__main__':
    n1 = Layer(2,1) # 設定   入力値が列、出力値が行
    bit_input = np.array([[[1]],
                          [[0]],
                          [[1]],
                          [[1]],
                          [[0]]])
    bit_output = np.array([[[1],
                            [0]],
                           [[0],
                            [1]],
                           [[1],
                            [0]],
                           [[1],
                            [0]],
                           [[0],
                            [1]]])
    for _ in range(10000):
        for i,o in zip(bit_input,bit_output):
            n1._y = n1.foward(i)
            delta = n1.back_propagation(o)
            update_y = n1.update(i)
#             print(back_parameters)
            error_ = np.power(update_y - o,2)
            print("誤差：", error_)
```
