```python
from tqdm import tqdm
```

```python
def numpy_time(n):  # y1 = [,,,,,,]
    total_time = []
    for i in tqdm(range(1,n,1)):
        x1 = np.random.rand(i) 
        x2 = np.random.rand(i)
        start_time = time()
        x1.dot(x2)
        end_time = time()
        total_time.append(end_time - start_time)
    return total_time
```

```python
def forloop_time(n):
    total_time = []
    for i in tqdm(range(1,n,1)):
        x1 = np.random.rand(i) 
        x2 = np.random.rand(i)
        result = 0
        start_time = time()
        for x,y in zip(x1,x2):
            result += x*y        
        end_time = time()
        total_time.append(end_time - start_time)
    return total_time
```

```python
x = np.arange(1,10000,1)
s = numpy_time(10000)
c = forloop_time(10000)
# plt.legend((x,c), ('$s$','$c$'))
numpy_plot, = plt.plot(x,s,'b-')
forloop_plot, = plt.plot(x,c, 'r-')
plt.legend((numpy_plot,forloop_plot), ('$numpyplot$','$forloopplot$'))
plt.xlabel('$x$')
plt.show()
```

```python
# 演習(距離の計算)
def distance(x1, x2):
    y = x2 - x1
    z = np.power(y, 2)
    return np.sqrt(z.sum())

x1 = np.array([0.,1.])
x2 = np.array([1.,0.])
distance(x1,x2)
```

```python
# n = 2 ~ 100
def dist(i, n):  
    total_time = []
    distance_list = []
    # x1_list = []
    # x2_list = []
    for j in tqdm(range(i,n,1)):
        x1 = power(2., j)  
        x2 = power(10., j)
        # x1_list.append(x1)
        # x2_list.append(x2)
        distance_list.append(distance(x1, x2))
        start_time = time()
        end_time = time()
        total_time.append(end_time - start_time)
    # return distance_list, x1_list, x2_list
    return distance_list

dist(2, 100)
```

```python
class Neuron:
    
    def __init__(self, n):
        self.w = np.random.rand(n)  # setされた次元数
        self.b = np.random.rand(1)  # [] 1次元の要素をバイアスに与える
        
    def set_state(self, x):
        s0 = (x.dot(self.w))+1*self.b
        return s0

    def activate(self, s0):
        y = 1 / (1 + np.exp(-s0))
        return y

    def output(self, x):
        return self.activate(self.set_state(x))

if __name__ == '__main__':
    n = Neuron(3) # 3次元に設定
    x = np.array([1., 2., 0.])
    print(n.output(x))
```
